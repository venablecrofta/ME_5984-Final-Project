```python
import numpy as np
import matplotlib.pyplot as plt

class BayesianLinearRegression:
    def __init__(self, alpha=1.0, beta=1.0):
        self.alpha = alpha  # Prior precision
        self.beta = beta    # Noise precision
        self.m_N = None     # Posterior mean
        self.S_N = None     # Posterior covariance

    def fit(self, X, y):
        N, D = X.shape
        I = np.eye(D)
        self.S_N = np.linalg.inv(self.alpha * I + self.beta * X.T @ X)
        self.m_N = self.beta * self.S_N @ X.T @ y

    def predict(self, X_test):
        mean = X_test @ self.m_N
        variance = (1 / self.beta) + np.sum(X_test @ self.S_N * X_test, axis=1)
        return mean, variance

# Example usage: Assuming df_train and df_test are your DataFrames

X_train = df_train[['s_2']].values  # double brackets to keep 2D
y_train = df_train['RUL'].values

X_test = df_test[['s_2']].values    # same for test
mean_preds_all = []
std_preds_all = []

# Instantiate the Bayesian Regression model
blr = BayesianLinearRegression()

# Fit Bayesian Regression for each engine (or repeat)
for engine_id in df_train['id'].unique():
    if engine_id % 2 == 0:  # Plot every other engine for better visualization
        # Extract data for the current engine
        df_engine = df_train[df_train['id'] == engine_id]
        X_engine = df_engine[['s_2']].values
        y_engine = df_engine['RUL'].values
        
        # Fit the Bayesian regression model
        blr.fit(X_engine, y_engine)
        
        # Predict the RUL using the trained model
        mean_pred, var_pred = blr.predict(X_engine)
        std_pred = np.sqrt(var_pred)
        
        # Ensure the predictions and uncertainties are 1D arrays
        mean_pred = np.ravel(mean_pred)
        std_pred = np.ravel(std_pred)

        # Store the predictions and uncertainties for averaging
        mean_preds_all.append(mean_pred)
        std_preds_all.append(std_pred)

        # Plot the predictions for each engine
        plt.plot(X_engine, mean_pred, label=f'Engine {engine_id} Mean Prediction', alpha=0.6)
        plt.fill_between(X_engine[:, 0], 
                         mean_pred - std_pred, 
                         mean_pred + std_pred, 
                         color='blue', alpha=0.3)

# Average all predictions and uncertainties
mean_preds_avg = np.mean(mean_preds_all, axis=0)
std_preds_avg = np.mean(std_preds_all, axis=0)

# Plot the averaged results
plt.plot(X_engine, mean_preds_avg, label='Average Mean Prediction', color='green', linewidth=2)
plt.fill_between(X_engine[:, 0], 
                 mean_preds_avg - std_preds_avg, 
                 mean_preds_avg + std_preds_avg, 
                 color='green', alpha=0.3, label='Average Uncertainty')

# Plot vertical line at RUL = 30
plt.axvline(30, color='red', linestyle='dashed', linewidth=2)

# Labels and aesthetics
plt.xlabel('Sensor 2 (s_2) value')
plt.ylabel('Remaining Useful Life (RUL)')
plt.title('Bayesian Linear Regression: Average RUL Prediction with Uncertainty')
plt.legend(loc='upper right', bbox_to_anchor=(1.15, 1))
plt.grid(True)
plt.show()
```
